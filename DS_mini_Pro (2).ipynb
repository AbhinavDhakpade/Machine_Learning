```python
k=k.drop(columns=['address'])
k=k.drop(columns=['name'])
k=k.drop(columns=['phone'])
k=k.drop(columns=['menu_item'])
k=k.drop(columns=['reviews_list'])
```

```python
k
```

```python
def new_rate(old_rate):
    if pd.isnull(old_rate):
        return np.nan
    else:
        return str(old_rate) and old_rate.split('/')[0]
```

```python
k['rate']=k['rate'].apply(new_rate)
```

```python
k
```

```python
k['rate'] = k['rate'].replace(['NEW', '-'], np.nan)
```

```python
k['rate'] = k['rate'].astype(float)
```

```python
k
```

```python
k.info()
```

```python
k['approx_cost(for two people)']=k['approx_cost(for two people)'].str.replace(',','').astype(float)
```

```python
k.info()
```

```python
k=pd.get_dummies(k,columns=['online_order','book_table','location','rest_type','dish_liked','cuisines','listed_in(type)','listed_in(city)'],drop_first=True)
```
