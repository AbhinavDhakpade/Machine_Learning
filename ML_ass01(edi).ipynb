```python
# importing two different datasets of white and red wine
r=pd.read_csv('winequality-red.csv',sep=';')
```

```python
w=pd.read_csv('winequality-white.csv',sep=';')
```

```python
r
```

```python
w
```

```python
r.isnull().sum()
```

```python
w.isnull().sum()
```

```python
#distinguish through visualization add separate labels to differnt datasets
r["wine_type"] = "red"
w["wine_type"] = "white"
```

```python
wine=pd.concat([r,w],axis=0)
```

```python
wine
```

```python
wine["new_wine"]=wine["wine_type"].map({'red':0,'white':1})
```

```python
wine
```

```python
#now i have a proper dataset so i shall eliminate column with categorical values and split the dataset for
#target and test data
x=wine.drop(columns=["wine_type","new_wine"])
y=wine['new_wine']
```

```python
x
```

```python
y
```

```python
x.isnull().sum()
```

```python
y.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rb=RobustScaler()
```

```python
x=rb.fit_transform(x)
```

```python
x
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_x=pca.fit_transform(x)
```

```python
plt.scatter(pca_x[:,0], pca_x[:,1], c=y, cmap='viridis', alpha=0.6)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA on Wine Dataset')
plt.colorbar(label='Wine Type')
plt.show()
```
