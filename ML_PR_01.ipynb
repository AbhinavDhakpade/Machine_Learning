```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk

```

```python
k=pd.read_csv('winequality.csv')
```

```python
k.info()
```

```python
k.isnull().sum()
```

```python
k.dropna(inplace=True)
```

```python
k
```

```python
k.isnull().sum()
```

```python
k['type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
k['type']=lb.fit_transform(k['type'])
```

```python
k
```

```python
x=k.drop('type',axis=1)
y=k['type']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
x_scal=sd.fit_transform(x)
```

```python
x_scal
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scal)
```

```python
x_pca
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.score
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x_pca[:,0],x_pca[:,1],cmap='plasma',c=k['type'])
plt.title("PCA")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cr)
print("Accuracy after PCA on Logistic Regression:",acc)
```
